```python
import yfinance as yf
import plotly.express as px

stock = yf.download("AAPL", start="2023-01-01", end="2024-01-01")

stock.columns = stock.columns.get_level_values(0)

fig = px.line(stock,
              x=stock.index,
              y='Close',
              title="Apple Stock Price")

fig.show()
```
